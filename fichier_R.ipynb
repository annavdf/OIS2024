```r
# Chargement des bibliothèques nécessaires
library(ggplot2)

# Génération d'un jeu de données factice sur la consommation d'eau
set.seed(123)
data <- data.frame(
  Consumption = rnorm(1000, mean = 150, sd = 30)  # 1000 valeurs aléatoires avec une moyenne de 150 et un écart-type de 30
)

# Création de l'histogramme
histogram <- ggplot(data, aes(x = Consumption)) +
  geom_histogram(binwidth = 5, fill = "skyblue", color = "black", alpha = 0.7) +
  labs(title = "Histogramme de la consommation d'eau",
       x = "Consommation d'eau (litres)",
       y = "Fréquence") +
  theme_minimal() +
  theme(
    plot.title = element_text(hjust = 0.5, size = 16, face = "bold"),
    axis.title = element_text(size = 14),
    axis.text = element_text(size = 12)
  )

# Affichage de l'histogramme
print(histogram)
```
